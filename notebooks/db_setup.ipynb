```python
import os
import sqlite3
import pandas as pd
from dotenv import load_dotenv
from nba_api.stats.static import teams
from nba_api.stats.endpoints import LeagueSeasonMatchups
import ace_tools_open as tools

load_dotenv()
```

```python
def database_connect():

    DB_PATH = os.getenv("DB_PATH", "nba_data.db")
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    return conn, cursor
```

```python
def get_teams():
    
    # NBA API: teams
    nba_teams = teams.get_teams()
    team_df = pd.DataFrame(nba_teams)
    team_df = team_df[['id', 'abbreviation', 'full_name']]

    return team_df
```

```python
# Create SQLite database
conn, cursor = database_connect()

# Create Teams table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Teams (
    team_id INTEGER PRIMARY KEY,
    abbreviation TEXT UNIQUE,
    full_name TEXT
);
""")

conn.commit()

# Retrieve NBA Teams
team_df = get_teams()


# Insert data into SQLite
team_df.to_sql("Teams", conn, if_exists="replace", index=False)
print("✅ Teams table created and populated successfully!")

# Query and display Teams table
query = "SELECT * FROM Teams;"
df = pd.read_sql(query, conn)

tools.display_dataframe_to_user(name="NBA Teams", dataframe=df)

# Close connection
conn.close()
```
