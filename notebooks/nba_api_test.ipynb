```python
import pandas as pd
from nba_api.stats.endpoints import leaguegamefinder

# Fetch all games for the current season
season = "2023-24"  # Modify this as needed
gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season)
games_df = gamefinder.get_data_frames()[0]

# Keep only relevant columns
games_df = games_df[["GAME_ID", "GAME_DATE", "TEAM_ID", "MATCHUP"]]

# Split MATCHUP column to get home/away teams
def extract_teams(row):
    parts = row["MATCHUP"].split(" ")
    if "vs." in row["MATCHUP"]:
        home_team, away_team = parts[0], parts[2]
    elif "@" in row["MATCHUP"]:
        away_team, home_team = parts[0], parts[2]
    return pd.Series({"home_team_abbr": home_team, "away_team_abbr": away_team})

games_df[["home_team_abbr", "away_team_abbr"]] = games_df.apply(extract_teams, axis=1)

# Merge with team IDs using the Teams table
conn = sqlite3.connect(DB_PATH)
teams_df = pd.read_sql("SELECT * FROM Teams;", conn)

# Create a mapping of abbreviations to team IDs
team_abbrev_to_id = dict(zip(teams_df["abbreviation"], teams_df["team_id"]))

# Map team abbreviations to team IDs
games_df["home_team_id"] = games_df["home_team_abbr"].map(team_abbrev_to_id)
games_df["away_team_id"] = games_df["away_team_abbr"].map(team_abbrev_to_id)

# Add season column
games_df["season"] = season

# Rename columns to match database schema
games_df.rename(columns={"GAME_ID": "game_id", "GAME_DATE": "game_date"}, inplace=True)

# Insert into SQLite
games_df.to_sql("Games", conn, if_exists="replace", index=False)
print("✅ Full season schedule inserted successfully!")

# Display table
import ace_tools as tools
tools.display_dataframe_to_user(name="NBA Full Season Schedule", dataframe=games_df)

conn.close()
```
